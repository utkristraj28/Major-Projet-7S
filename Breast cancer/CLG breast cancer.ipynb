# Data Cleaning

```python
df.isnull().sum()
```

# Scaling of data

there is no null data.

So, now we scale the data for better prediction

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scaled = scale.fit_transform(df.drop('target',axis=1))
```

```python
x = scaled
y = df['target']
```

# Spliting Data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=101)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis = LogisticRegression()
```

```python
logis.fit(xtrain,ytrain)
```

```python
pred_logis = logis.predict(xtest)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(pred_logis,ytest))
print('\n')
print(confusion_matrix(pred_logis,ytest))
```

# Simple Vector Machine

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(xtrain,ytrain)
```

```python
pred_svc = svc.predict(xtest)
```

```python
print(classification_report(pred_svc,ytest))
print('\n')
print(confusion_matrix(pred_svc,ytest))
```

# K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    pred_knn = knn.predict(xtest)
    error.append(np.mean(pred_knn != ytest))
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,40),error,ls='--',marker='o')
```

```python
knn1 = KNeighborsClassifier(n_neighbors=6)
```

```python
knn1.fit(xtrain,ytrain)
```

```python
pred_knn1 = knn1.predict(xtest)
```

```python
print(classification_report(pred_knn1,ytest))
print('\n')
print(confusion_matrix(pred_knn1,ytest))
```
